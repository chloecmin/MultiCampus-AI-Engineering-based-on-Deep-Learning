```python
def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/"+word
    return url
```

```python
#κ²μ‹κΈ€ λ‹¤μ–‘ν• μ •λ³΄ 
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    
    # κ²μ‹κΈ€ μ½κΈ°
    try:
        content = soup.select("div.C4VMK > span")[0].text
    except:
        content = "" 

    ## contentμ—μ„ ν•΄μ‹νƒκ·Έλ§ κ°€μ Έμ¤κΈ°
    hashtag =re.findall("#[\w,κ°€-ν£γ„±-γ…]+", content) 
    
    #λ‚ μ§ κ°€μ Έμ¤κΈ°
    date = soup.select("time._1o9PC.Nzb55")[0]['datetime'][:10]
    
    #μΆ‹μ•„μ” μ κ°€μ Έμ¤κΈ°
    like = soup.select("div.Nm9Fw > button > span")[0].text
    
    #μ„μΉ μ •λ³΄ κ°€μ Έμ¤κΈ°
    place = soup.select("div.M30cS")[0].text
        
    data = [content, date, like, place, hashtag]
    return data
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
browser = webdriver.Chrome("C://mychrome/chromedriver.exe")
```

```python
url = "https://www.starbucks.co.kr/store/store_map.do?disp=locale"
browser.get(url)
```

```python
#μ„μΈμ‹ λ²„νΌ
seoulBtn = "div.loca_step1_cont > ul > li:nth-child(1) > a"
browser.find_element_by_css_selector(seoulBtn).click()
```

```python
#μ„μΈμ‹ > μ „μ²΄λ²„νΌ
allBtn = "#mCSB_2_container > ul > li:nth-child(1) > a"
browser.find_element_by_css_selector(allBtn).click()
```

```python
html = browser.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
stores = soup.select("li.quickResultLstCon")

stores_li = []
for store in stores:
    name = store.select("strong")[0].text.strip()
    lat = store['data-lat'].strip()
    lng = store['data-long'].strip()
    store_type = store.select("i")[0]['class'][0][4:]
    
    addr_tel = str(store.select("p.result_details")[0]).split('<br/>')
    addr = addr_tel[0].split('>')[1]
    tel = addr_tel[1].split('<')[0]

    stores_li.append([name, lat, lng, store_type, addr, tel])
stores_li
```

```python
def seoul_open_api_data(url, service):
    data_list = None
    try : 
        result_dict = requests.get(url).json()
        result_data = result_dict[service]
        code = result_data['RESULT']["CODE"]
        if code == 'INFO-000':
            data_list = result_data['row']
    except Exception as e:
        print(e)
    return data_list
```

```python
#μ¤νƒ€λ²…μ¤ dfμ—μ„ μ£Όμ†μ»¬λΌλ§ μ¶”μ¶
sgg_names = []
for addr in seoul_starbucks['μ£Όμ†']:
    sgg = addr.split()[1]
    sgg_names.append(sgg)

seoul_starbucks['μ‹κµ°κµ¬λ…'] = sgg_names
seoul_starbucks.head()

seoul_starbucks.to_excel("seoul_starbucks_list.xlsx", index=False)
```

```python
#λ°μ΄ν„° λ³‘ν•©
```

```python
#μ¤νƒ€λ²…μ¤ λ°μ΄ν„°(λ§¤μ¥λ… | μ„λ„ | κ²½λ„ | λ§¤μ¥νƒ€μ… | μ£Όμ† | μ „ν™”λ²νΈ)
seoul_starbucks = pd.read_excel("seoul_starbucks_list.xlsx")
seoul_starbucks
```

```python
#κµ¬ λ³„λ΅ μ¤νƒ€λ²…μ¤ λ§¤μ¥ κ°μ μ„ΈκΈ° (ν”Όλ΄‡ν•¨μ μ΄μ©)
"""
ν”Όλ΄‡(pivot)ν•¨μ : ν”Όλ΄‡ν…(νΉμ • κΈ°μ¤€μ— λ”°λ¥Έ λ°μ΄ν„° μ”μ•½)
############## μ #############################
           μ¤νƒ€λ²…μ¤ λ§¤μ¥ μ
μ‹κµ°κµ¬λ…  |     
κ°•λ‚¨κµ¬    |      200
"""
starbucks_sgg_count = seoul_starbucks.pivot_table(index='μ‹κµ°κµ¬λ…', values="λ§¤μ¥λ…", aggfunc='count').rename(columns={
    'λ§¤μ¥λ…' : 'μ¤νƒ€λ²…μ¤_λ§¤μ¥μ'
}) #aggfunc : μ§‘κ³„ν•¨μ
```

```python
starbucks_sgg_count = starbucks_sgg_count.drop(['67'])
starbucks_sgg_count
```

```python
#μ‹κµ°κµ¬ λ©λ΅ λ°μ΄ν„° (μ‹κµ°κµ¬ μ½”λ“ | μ‹κµ°κµ¬λ… | μ„λ„ | κ²½λ„)
seoul_sgg = pd.read_excel('seoul_sgg_list.xlsx')
```

```python
# merge() : λ‘κ°μ λ°μ΄ν„° ν”„λ μ„μ„ ν•©μΉλ” ν•¨μ
# howλ” μ–΄λμ½ joinν•  κ²ƒμΈμ§€. μ™Όμ½μ΄ κΈ°μ¤€μ΄ λμ–΄ ν•©μΉκΈ°(merge)
seoul_sgg_final = pd.merge(seoul_sgg, starbucks_sgg_count, how='left', on='μ‹κµ°κµ¬λ…')
seoul_sgg_final
```

```python
#μ‹κµ°κµ¬λ³„ μΈκµ¬ν†µκ³„ μ—‘μ…€ λ¶λ¬μ¤κΈ°
seoul_sgg_pop = pd.read_excel("sgg_pop.xlsx")
```

```python
seoul_sgg_final = pd.merge(seoul_sgg_final, seoul_sgg_pop, how='left', on='μ‹κµ°κµ¬λ…')
seoul_sgg_final
```

```python
# μ„μΈμ‹ μ‚¬μ—…μ²΄ λ°μ΄ν„° μ—‘μ…€νμΌ λ¶λ¬μ¤κΈ°
seoul_sgg_biz = pd.read_excel("sgg_biz.xlsx")
```
